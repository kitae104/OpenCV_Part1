```python
import cv2
vidcap = cv2.VideoCapture('D:\Github\Vision_WS\OpenCV_Part1/videos\son_01.mp4')
success,image = vidcap.read()
count = 0
while success:
  cv2.imwrite(f"D:\Github\Vision_WS\OpenCV_Part1\outputs\son_frames/son_{count:06d}.jpg", image)     # save frame as JPEG file
  success,image = vidcap.read()
  print('새 프레임 읽기 : ', success)
  count += 1

print("변경 완료! (convert video to frame)")
```
