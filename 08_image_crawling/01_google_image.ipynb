```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import os
import time
import urllib.request

# 웹 드라이버 서비스 생성
service = Service(ChromeDriverManager().install())

# 검색할 키워드
search_keyword = 'Tedy Bruschi' # '손흥민' #input("어떤 이미지를 검색할까요? ") #'blackpink'

# 이미지 저장 폴더 경로
save_directory = 'images/' + search_keyword + '/'

# images 폴더가 없을 경우 폴더 생성
if not os.path.exists(save_directory):
    os.makedirs(save_directory)

# 웹 드라이버 실행
driver = webdriver.Chrome(service=service)

# Google 이미지 검색 페이지 열기
driver.get("https://www.google.co.kr/search?q=" + search_keyword + "&tbm=isch")

# 이미지 로딩을 위해 잠시 대기
time.sleep(5)

# 이미지 클릭하고 큰 이미지 저장하기
images = driver.find_elements(By.CSS_SELECTOR, ".rg_i")
i = 1
for image in images:
    if i == 51:
        break
      
    try:
        # 이미지 클릭
        driver.execute_script("arguments[0].click();", image)
        
        # 이미지 로딩을 위해 잠시 대기
        time.sleep(2)
        
        # 큰 이미지 URL 가져오기
        big_image = driver.find_element(By.CSS_SELECTOR, "#Sva75c > div.A8mJGd.NDuZHe.CMiV2d.OGftbe-N7Eqid-H9tDt > div.dFMRD > div.AQyBn > div.tvh9oe.BIB1wf.hVa2Fd > c-wiz > div > div > div > div > div.v6bUne > div.p7sI2.PUxBg > a > img.sFlh5c.pT0Scc.iPVvYb")
        # big_image = driver.find_element(By.XPATH, "/html/body/div[2]/c-wiz/div[3]/div[2]/div[3]/div[2]/div[2]/div[2]/div[2]/c-wiz/div/div/div/div[3]/div[1]/a/img[1]")
        image_url = big_image.get_attribute("src")
        
        # 이미지 저장
        file_name = save_directory + f"{i}.jpg"
        urllib.request.urlretrieve(image_url, file_name)
        print(f"이미지 저장 완료: {file_name}")
        
        # 뒤로 가기
        driver.execute_script("window.history.go(-1)")
        
        # 이미지 로딩을 위해 잠시 대기
        i += 1
        time.sleep(2)
    except Exception as e:        
        print(f"이미지 저장 실패: {e}")

# 웹 드라이버 종료
driver.quit()
```
